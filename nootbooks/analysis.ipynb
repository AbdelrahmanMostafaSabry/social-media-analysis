```python
plt.figure(figsize=(8,5))

sns.histplot(data=df, x="age", bins=7)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Number of Users")

plt.show()
```

### Insight 1

The participants' ages are distributed fairly evenly between 13 and 19 years, with no significant concentration at a specific age.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="platform_usage"
)

plt.title("Platform Usage Distribution")
plt.xlabel("Platform")
plt.ylabel("Number of Users")

plt.show()
```

```python
df["platform_usage"].value_counts()
```

### Insight 2

Instagram is the most frequently used platform among the participants. However, the difference between Instagram, TikTok, and users who use both platforms is relatively small, indicating that platform usage is fairly balanced across the dataset.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="daily_social_media_hours",
    y="sleep_hours",
    alpha=0.5
)

plt.title("Daily Social Media Hours vs Sleep Hours")
plt.xlabel("Daily Social Media Hours")
plt.ylabel("Sleep Hours")

plt.show()
```

```python
df["daily_social_media_hours"].corr(df["sleep_hours"])
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="daily_social_media_hours",
    y="stress_level",
    alpha=0.5
)

plt.title("Daily Social Media Hours vs stress level")
plt.xlabel("Daily Social Media Hours")
plt.ylabel("stress level")

plt.show()
```

```python
df["daily_social_media_hours"].corr(df["stress_level"])
```

```python
correlation=df.corr(numeric_only=True)
correlation
```

```python
plt.Figure(figsize=(10,8))
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation.unstack().sort_values(ascending=False)
```

```python
corr_matrix = df.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

corr_matrix = corr_matrix.mask(mask)

top_corr = (
    corr_matrix
    .unstack()
    .dropna()
    .sort_values(ascending=False)
)

top_corr
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="depression_label",
    y="daily_social_media_hours"
)

plt.title("Daily Social Media Hours by Depression Status")

plt.xlabel("Depression Label")

plt.ylabel("Daily Social Media Hours")

plt.show()
```

Participants labeled with depression generally spend more time on social media than participants without depression.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="depression_label",
    y="sleep_hours"
)

plt.title("Sleep Hours by Depression Status")

plt.xlabel("Depression Label")

plt.ylabel("Sleep Hours")

plt.show()
```

Participants with depression generally sleep fewer hours than participants without depression.

```python
# ============================
# Insight 5
# Platform Usage vs Daily Social Media Hours
# Library: Seaborn
# ============================

plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="platform_usage",
    y="daily_social_media_hours"
)

plt.title("Daily Social Media Hours by Platform")

plt.xlabel("Platform")

plt.ylabel("Daily Social Media Hours")

plt.show()
```

Daily social media usage is very similar across Instagram, TikTok, and users who use both platforms. No platform shows a significantly higher usage duration.

```python
# =====================================
# Insight 6
# Depression Label Distribution
# Library: Seaborn
# =====================================

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="depression_label"
)

plt.title("Depression Label Distribution")

plt.xlabel("Depression Label")

plt.ylabel("Number of Participants")

plt.show()
```

```python
# ==========================================
# Dashboard Layout
# Library: Matplotlib
# ==========================================

fig, axes = plt.subplots(
    nrows=2,
    ncols=3,
    figsize=(18,10)
)

plt.tight_layout()
# ==========================================
# Chart 1
# Age Distribution
# Library: Seaborn
# ==========================================

sns.histplot(
    data=df,
    x="age",
    bins=7,
    ax=axes[0,0]
)

axes[0,0].set_title("Age Distribution")

axes[0,0].set_xlabel("Age")

axes[0,0].set_ylabel("Number of Users")
# ==========================================
# Chart 2
# Platform Distribution
# Library: Seaborn
# ==========================================

sns.countplot(
    data=df,
    x="platform_usage",
    ax=axes[0,1]
)

axes[0,1].set_title("Platform Usage")

axes[0,1].set_xlabel("Platform")

axes[0,1].set_ylabel("Users")
# ==========================================
# Chart 3
# Depression Distribution
# Library: Seaborn
# ==========================================

sns.countplot(
    data=df,
    x="depression_label",
    ax=axes[0,2]
)

axes[0,2].set_title("Depression Distribution")

axes[0,2].set_xlabel("Depression")

axes[0,2].set_ylabel("Participants")

# ==========================================
# Chart 4
# Daily Social Media Hours vs Depression
# Library: Seaborn
# ==========================================

sns.boxplot(
    data=df,
    x="depression_label",
    y="daily_social_media_hours",
    ax=axes[1,0]
)

axes[1,0].set_title("Social Media Hours vs Depression")

axes[1,0].set_xlabel("Depression Label")

axes[1,0].set_ylabel("Hours")

# ==========================================
# Chart 5
# Sleep Hours vs Depression
# Library: Seaborn
# ==========================================

sns.boxplot(
    data=df,
    x="depression_label",
    y="sleep_hours",
    ax=axes[1,1]
)

axes[1,1].set_title("Sleep Hours vs Depression")

axes[1,1].set_xlabel("Depression Label")

axes[1,1].set_ylabel("Sleep Hours")

# ==========================================
# Chart 6
# Daily Social Media Hours by Platform
# Library: Seaborn
# ==========================================

sns.boxplot(
    data=df,
    x="platform_usage",
    y="daily_social_media_hours",
    ax=axes[1,2]
)

axes[1,2].set_title("Hours by Platform")

axes[1,2].set_xlabel("Platform")

axes[1,2].set_ylabel("Hours")
fig.suptitle(
    "Social Media Impact Analysis Dashboard",
    fontsize=20,
    fontweight="bold"
)
plt.tight_layout(rect=[0,0,1,0.96])
sns.set_theme(style="whitegrid")
plt.savefig(
    "../images/dashboard.png",
    dpi=300,
    bbox_inches="tight"
)
plt.show()
```
